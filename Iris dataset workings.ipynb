```python
import pandas as pd
import numpy as np

# Load dataset
iris_data = pd.read_csv(r"C:\Users\Asus\Downloads\iris dataset\Iris.csv")

# Add a new column with a value to calculate sepal length
iris_data["sepal_area"] = iris_data["sepal_length"] * iris_data["sepal_width"] 

# Add a new column with a value to calculate petal length
iris_data["petal_area"] = iris_data["petal_length"] * iris_data["petal_width"] 

# Add a new column with a value to calculate ratio
iris_data["ratio"] = iris_data["sepal_area"] / iris_data["petal_area"]

# Display updated Data
print(iris_data.head())
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Compute correlation matrix 
correlation_matrix = iris_data.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
